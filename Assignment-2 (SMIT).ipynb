# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks=int(input("Enter Your Marks:"))

if marks >= 90 :
    print("A")
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
elif marks >= 60:
    print("D")
else:
    print("Fail")


    
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day=int(input("Enter your number (1-7) :"))

if day == 1:
    print("MONDAY")
elif day == 2:
    print("TUESDAY")
elif day == 3:
    print("WEDNESDAY")
elif day == 4:
    print("THURSDAY")
elif day == 5:
    print("FRIDAY")
elif day == 6:
    print("SATURDAY")
elif day == 7:
    print("SUNDAY")
else:
    print("Value you entered is incorrect !")

    
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month=int(input("Enter Number Of Your Choice (1-12):"))

if month == 12 or month == 1 or month == 2:
    print("Winter")
elif month == 3 or month == 4 or month == 5:
    print("Spring")
elif month == 6 or month == 7 or month == 8:
    print("Summer")
else:
    print("Autumn")
      
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"
# pehla < dusra → "Second is greater"
# warna → "Both equal"

```python
pehla=float(input("Enter your 1st number:"))
dusra=float(input("Enter you 2nd number:"))

if pehla > dusra:
    print("First is Greater")
elif pehla < dusra:
    print("Second is Greater")
else:
    print("Both equal")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
operator = input("Enter Operator Of Your Choice:")

if operator == "+" :
    print("Addition")
elif operator == "-":
    print("Subtraction")
elif operator == "*":
    print("Multiplication")
elif operator == "/":
    print("Division")
else:
    print("Invalid Operator")
```
